# Plotting

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exercise 1

Import the company_sales_data.csv dataset which can be found in the data directory of this course. 
This data contains the monthly sales for each product for a specific year of a specific company. 
Write a program that plots the monthly evolution of sales for each product.
The plot should look like this:

<img src="../data/img/plot_1.png" width="400">

```python
# define the path of the datafile
data_path = "../data/company_sales_data.csv"

# import data
df = pd.read_csv(data_path)
```

```python
# define the product variables
product_vars = ["facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]

# initialize the figure
plt.figure(figsize=(8, 6))
# loop through the products
for product in product_vars:
    
    # plot the evolution of sales of product
    plt.plot(df["month_number"], df[product], label=product)

# add labels
plt.xlabel("Month")
plt.ylabel("Sales")
# add legend
plt.legend(loc="upper left")
# add title
plt.title("Monthly Sales of Products")
# show plot
plt.show()
```

## Exercise 2

Import the company_sales_data.csv dataset which can be found in the data directory of this course. 
This data contains the monthly sales for each product for a specific year of a specific company. 
Create a pie chart that shows how much each product contributed to the total sales of the company.
The plot should look like this:

<img src="../data/img/plot_2.png" width="400">

```python
# define the path of the datafile
data_path = "../data/company_sales_data.csv"

# import data
df = pd.read_csv(data_path)
```

```python
# get the product names
product_vars = ["facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]

# get the total sales per product
total_sales_per_product = [df["facewash"].sum(), 
                           df["toothpaste"].sum(), 
                           df["bathingsoap"].sum(), 
                           df["shampoo"].sum(),
                           df["moisturizer"].sum()]


# create pie chart
plt.figure(figsize=(6, 6))
plt.pie(total_sales_per_product, labels=product_vars)
plt.show()

```
